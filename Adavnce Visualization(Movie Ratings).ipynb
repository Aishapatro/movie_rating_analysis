```python
sns.set_style('darkgrid')

f, axes = plt.subplots(2, 2, figsize=(15, 15))

# Top-left
k1 = sns.kdeplot(data=movies, x='BudgetMillions', y='AudienceRating', ax=axes[0, 0],fill=True)

# Top-right
k2 = sns.kdeplot( data=movies, x='BudgetMillions', y='CriticaRating', ax=axes[0, 1], fill=True)

# Set limits
k1.set(xlim=(-20, 160))
k2.set(xlim=(-20, 160))

# Bottom-left
z = sns.violinplot( data=movies[movies.Genre == 'Drama'], x='Year', y='CriticaRating', ax=axes[1, 0])

# Bottom-right
k4 = sns.kdeplot(  data=movies,  x='CriticaRating',  y='AudienceRating',  cmap='Reds',  fill=True,ax=axes[1, 1])

k4b = sns.kdeplot(data=movies,x='CriticaRating',y='AudienceRating',cmap='Reds',ax=axes[1, 1])

plt.show()
```

```python
sns.set_style('dark', {'axes.facecolor': 'black'})

f, axes = plt.subplots(2, 2, figsize=(15, 15))

# plot [0,0]
k1 = sns.kdeplot(data=movies,x='BudgetMillions',y='AudienceRating',fill=True,cmap='inferno',ax=axes[0, 0])

k1b = sns.kdeplot(data=movies,x='BudgetMillions',y='AudienceRating',cmap='cool',  ax=axes[0, 0])

# plot [0,1]
k2 = sns.kdeplot(data=movies, x='BudgetMillions', y='CriticaRating', fill=True, cmap='inferno', ax=axes[0, 1])

k2b = sns.kdeplot( data=movies, x='BudgetMillions',y='CriticaRating', cmap='cool', ax=axes[0, 1])

# plot [1,0]
z = sns.violinplot( data=movies[movies.Genre == 'Drama'], x='Year', y='CriticaRating', ax=axes[1, 0])

# plot [1,1]
k4 = sns.kdeplot(data=movies,x='CriticaRating',y='AudienceRating', fill=True, cmap='Blues_r', ax=axes[1, 1])

k4b = sns.kdeplot( data=movies, x='CriticaRating', y='AudienceRating', cmap='gist_gray_r', ax=axes[1, 1])

# axis limits
k1.set(xlim=(-20, 160))
k2.set(xlim=(-20, 160))

plt.show()
```
